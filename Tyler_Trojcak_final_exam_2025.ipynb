* If `twoorthree` returns a `3`:

````python
# your takeshot function here
def takeshot():
  shot_type = twoorthree()
  if shot_type == 2:
    scoreornot2 = twopoint()
    if scoreornot2 == True:
      return 2
    else:
      return 0
  if shot_type == 3:
    scoreornot3 = threepoint()
    if scoreornot3 == True:
      return 3
    else:
      return 0
````

(10 points)

* Use the Euler method to determine the vertical position and velocity of the rocket as a function of time, until the rocket reaches an altitude of 9000 meters, higher than the highest mountain on Earth.

* You will need to create a while loop and continue the calculate until the y position exceeds 9000 m.  (NOTE: be sure to increment your counter after each pass through the while loop!)


* Assume the following values for air resistance:
  * $B_2 = 0.1$

* Assume the following initial conditions:
  * $y_0 = 2$ m
  * $v_{0y} = 10$ m/s


````python
# your code here
B2 = 0.1
y0 = 2 # m
yf = 9000
v0y = 10 # m/s
dt = 0.1

ay = [0]
v = [v0y]
y = [y0]
t = [0]

while y[-1] < yf:
  ay.append(10.2 - (B2*v[-1]**2)/1000)
  v.append(v[-1] + ay[-1]*dt)
  y.append(y[-1] + v[-1]*dt)
  t.append(t[-1] + dt)
````

## Problem 3: Superposition of Waves

### Part (3a)

(5 points)

* Create a plot of that shows the sum of two cosine waves of the form:
$$y(t) = A cos (\omega t + \phi) $$

  * The first wave has an amplitude of 1, an angular velocity of $2\pi$, and a phase constant of $\pi$.

  * The second wave has an amplitude of 1.5, an angular velocity of $4\pi$, and a phase constant of zero.

* The x axis of the plot should range from $0 < t < 2$ sec.

* Call the sum of the two waves `ysum`.

````python
# your code here
t = np.linspace(0, 2, 100)
wave1 = 1*np.cos(2*np.pi*t + np.pi)
wave2 = 1.5*np.cos(4*np.pi*t + 0)

ysum = wave1 + wave2

plt.plot(t, ysum, label="summed wave")
plt.plot(t, wave1, label="wave1")
plt.plot(t, wave2, label="wave2")
plt.legend()
````

### Part (3b)

(5 points)

* Take the FFT of `ysum`.

* To do this, use the numpy function `np.fft.rfft`.  This will return the amplitude of the Fourier series.

* You can compute the frequency for the Fourier transform using `np.fft.rfftfreq`.  You should call this as

```
dt = time[1]-time[0]
freq = np.fft.rfftfreq(len(ysum),d=dt)
```

* Graph the absolute value of the amplitude of the Fourier series versus frequency.


````python
# find fft of ysum and the abs of it
ysumfft = np.fft.rfft(ysum)
psumfft = np.abs(ysumfft)

dt = t[1] - t[0]
freq = np.fft.rfftfreq(len(ysum),d=dt)

# plot abs value and frequency
plt.plot(freq, psumfft)
````
