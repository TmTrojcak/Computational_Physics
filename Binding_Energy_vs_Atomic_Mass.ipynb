```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data
file_path = "/binding.txt"
data = pd.read_csv(file_path, delim_whitespace=True, header=None, names=["Atomic Number", "Element", "Binding Energy", "Mass Number"])
```

```python
data
```

```python
# Extract relevant columns for plotting
mass_numbers = data["Mass Number"]
binding_energies = data["Binding Energy"]
```

```python
# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(mass_numbers, binding_energies, marker='o', linestyle='-', color='blue')

# Add labels and title
plt.title('Binding Energy per Nucleon vs. Mass Number', fontsize=14)
plt.xlabel('Mass Number', fontsize=12)
plt.ylabel('Binding Energy per Nucleon (MeV)', fontsize=12)

# Show gridlines
plt.grid()

# Display the plot
plt.show()
```
